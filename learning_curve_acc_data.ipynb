```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit
from utils.data_storage import DataContianer
from utils.data_visualization_utils import *
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
drone1 = "drone1"
drone2 = "drone2"
drone3 = "drone3"
drone4 = "drone4"
title_rf='Learning Curve (Random Forest)'
title_adb='Learning Curve (AdaBoost)'
```

```python
cv = ShuffleSplit(n_splits=10, test_size=.25, random_state=0)
```

# Drone 1

```python
drone1_acc_data = DataContianer(drone1, 'acc', 4)
```

```python
X_train_d1 = drone1_acc_data.X_train_transformed
y_train_d1 = drone1_acc_data.y_train_transformed
X_test_d1 = drone1_acc_data.X_test_transformed
y_test_d1 = drone1_acc_data.y_test_transformed
```

# Random Forest

```python
rf_d1 = RandomForestClassifier(n_estimators=100, random_state=0)
rf_d1.fit(X_train_d1, y_train_d1)
```

```python
plot_learning_curve(rf_d1, title_rf, X_train_d1, y_train_d1, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5))
plt.show()
```

# AdaBoost

```python
adb_d1 = AdaBoostClassifier(DecisionTreeClassifier(max_depth=2),
                         algorithm="SAMME",
                         n_estimators=100, learning_rate=0.45)

adb_d1.fit(X_train_d1, y_train_d1)
```

```python
plot_learning_curve(adb_d1, title_adb, X_train_d1, y_train_d1, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5))
plt.show()
```

```python
adb_d1.score(X_test_d1, y_test_d1)
```
